```python
lb=q1-1.5*IQR
ub=q3+1.5*IQR
```

```python
df.loc[(df["Item_Visibility"] > ub) | (df["Item_Visibility"] < lb)].head(3)
```

```python
df1=df.loc[(df["Item_Visibility"] < ub) & (df["Item_Visibility"] > lb)]
print("before removing outliers:",len(df))
print("after removing outliers:",len(df1))
print("outliers:",len(df)-len(df1))
```

```python
df1=df.copy()
df1.loc[(df1["Item_Visibility"]>ub), "Item_Visibility"]=ub
df1.loc[(df1["Item_Visibility"]<lb), "Item_Visibility"]=lb
```

# Split data into training and testing.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# import Linear Regression model and its evaluation metrics.

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error,r2_score
```

```python
lr=LinearRegression()
lr=lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
pred=lr.predict(X_test)
```

```python
print("MSE:",metrics.mean_squared_error(y_test,pred))
print("---"*21)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,pred)))
print("---"*21)
print("r2_SCORE:",metrics.r2_score(y_test,pred))
```
